```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from keras.models import Sequential
from keras.layers import GRU, Dense
from keras.optimizers import Adam

# Step 1: Load and preprocess data
# Load the dataset
data = pd.read_csv("/content/ABBANK.csv")

# Convert 'Date' column to datetime format and set it as the index
data['Date'] = pd.to_datetime(data['Date'], format='%m/%d/%Y')
data.set_index('Date', inplace=True)

# Check for missing values
print(data.isnull().sum())  # No missing values in the data

# Step 2: ARIMA - Check for stationarity
# Perform the Augmented Dickey-Fuller test for stationarity
adf_result = adfuller(data['Close'])
print(f"ADF test p-value: {adf_result[1]}")

# Step 3: Differencing to make data stationary (if necessary)
# Differencing the data
data['Close_diff'] = data['Close'].diff().dropna()

# Perform ADF test again
adf_result_diff = adfuller(data['Close_diff'].dropna())
print(f"ADF test p-value after differencing: {adf_result_diff[1]}")

# Step 4: ACF and PACF plots
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.figure(figsize=(12, 6))

# ACF plot
plt.subplot(1, 2, 1)
plot_acf(data['Close_diff'].dropna(), lags=30, ax=plt.gca())

# PACF plot
plt.subplot(1, 2, 2)
plot_pacf(data['Close_diff'].dropna(), lags=30, ax=plt.gca())

# Save the figure
plt.savefig('acf_pacf_plots.png', dpi=300)

# Show the figure
plt.show()


# Step 5: ARIMA Model
# Fit ARIMA model (p=1, d=1, q=1)
arima_model = ARIMA(data['Close'], order=(1, 1, 1))
arima_fit = arima_model.fit()

# Print ARIMA model summary
print(arima_fit.summary())

# Step 6: Prepare data for GRU Model
# Normalize the 'Close' prices for neural network
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data[['Close']])

# Prepare data for GRU - Create sequences of past values to predict the next value
def create_sequences(data, seq_length):
    sequences = []
    labels = []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i+seq_length])
        labels.append(data[i+seq_length])
    return np.array(sequences), np.array(labels)

# Create sequences and labels (e.g., using past 60 days to predict next day's price)
seq_length = 60
X, y = create_sequences(data_scaled, seq_length)

# Split into training and testing sets (80% train, 20% test)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Reshape for GRU input: [samples, time steps, features]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Step 7: Build and Train the GRU Model
gru_model = Sequential()
gru_model.add(GRU(units=50, return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])))
gru_model.add(Dense(units=1))  # Output layer to predict the next day's price

# Compile the model
gru_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Train the model
gru_model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Step 8: Evaluate the GRU Model
predicted_price = gru_model.predict(X_test)

# Inverse transform the predicted prices to get actual scale
predicted_price = scaler.inverse_transform(predicted_price)

# Evaluate performance (Plotting)
plt.figure(figsize=(10, 6))
plt.plot(data.index[-len(y_test):], scaler.inverse_transform(y_test.reshape(-1, 1)), label='Actual Price')
plt.plot(data.index[-len(y_test):], predicted_price, label='Predicted Price')
plt.title('Stock Price Prediction - GRU vs Actual')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Step 9: Evaluate ARIMA Model Performance (Prediction)
arima_pred = arima_fit.forecast(steps=len(y_test))

# Plot ARIMA predictions against actual values
plt.figure(figsize=(10, 6))
plt.plot(data.index[-len(y_test):], scaler.inverse_transform(y_test.reshape(-1, 1)), label='Actual Price')
plt.plot(data.index[-len(y_test):], arima_pred, label='ARIMA Predicted Price')
plt.title('Stock Price Prediction - ARIMA vs Actual')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
